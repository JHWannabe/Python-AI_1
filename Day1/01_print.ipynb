'%'    

```python
#  
# d(), f(), s() '%'       .
#         .
#   
# %[-][0][n][.m] => []    . =>  
# n:    
# -:         .
# 0:          '0'  .
# .m:   
```

```python
print(100)
print(1)
print(10000)
```

```python
# print(' ' %  )
print('%5d' % 100)
print('%-5d' % 100)
print('%05d' % 100)
print('%5d' % 1000)
print('%5d' % 10000)
```

```python
print('abc')
print('a')
print('abcde')
```

```python
print('%5s' % 'abc')
print('%5s' % 'a')
print('%5s' % 'abcde')
print('%-5s' % 'a')
print('%05s' % 'a') #    '0' .
```

```python
print('%6.2f' % 123.45)
print('%6.2f' % 123.4)   #     '0'  .
print('%6.2f' % 123.456) #      . =>    .
```

```python
#      2  '%'    ()  .
print('%d + %d = %d' % (5, 3, 5 + 3)) # 
print('%d - %d = %d' % (5, 3, 5 - 3)) # 
print('%d * %d = %d' % (5, 3, 5 * 3)) # 
#  '/'     .
print('%d / %d = %d' % (5, 3, 5 / 3)) # ,   %d      1 .
print('%d / %d = %f' % (5, 3, 5 / 3)) # 
#  '//'    .
print('%d // %d = %d' % (5, 3, 5 // 3)) # 
print('%d // %d = %f' % (5, 3, 5 // 3))
# '%'      '%'     '%%' 2  .
print('%d %% %d = %f' % (5, 3, 5 % 3)) # 
print('%d %d = %d' % (2, 10, 2 ** 10)) # 
print(' %d = %f' % (2, 2 ** 0.5)) # 
```

  : ()   => ** => *, /, //, % => +, -

    format()  {}   

```python
print('%d + %d = %d' % (5, 3, 5 + 3))
print('{} + {} = {}' .format(5, 3, 5 + 3))
print('{} - {} = {}' .format(5, 3, 5 - 3))
print('{} * {} = {}' .format(5, 3, 5 * 3))
print('{} / {} = {}' .format(5, 3, 5 / 3))
print('{} // {} = {}' .format(5, 3, 5 // 3))
print('{} % {} = {}' .format(5, 3, 5 % 3))
print('{} ** {} = {}' .format(5, 3, 5 ** 3))
```

  format()      

```python
print('%s %d.' % ('', 20))
print('{} {}.' .format('', 20))
print('{0} {1}.' .format('', 20))
```

```python
# {:} =>    '%'  .
print('{0:10s} {1:5d}.' .format('', 20))
print('{0:10s} {1:7,d}.' .format('', 2000000)) #   ',' .
print('{0:10s} {1:5.2f}.' .format('', 20.456))
```

'<', '>', '^'        . =>  '<'  .

```python
print('{0:10s} {1:5d}.' .format('', 20))
print('{0:<10s} {1:5d}.' .format('', 20))
print('{0:>10s} {1:5d}.' .format('', 20))
print('{0:^10s} {1:5d}.' .format('', 20))
```

'<', '>', '^'           .

```python
print('{0:_<10s} {1:5d}.' .format('', 20))
print('{0:#>10s} {1:5d}.' .format('', 20))
print('{0:*^10s} {1:5d}.' .format('', 20))
```
